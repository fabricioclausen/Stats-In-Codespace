```python
################################################################################
#     ANÁLISES COMPARATIVAS ENTRE 2020 E 2021 - PREPARAÇÃO DO DATASET          #
################################################################################

# Carregar bibliotecas necessárias
library(tidyverse)
library(kableExtra)

################################################################################
#              PADRONIZAR E UNIFICAR DATASETS 2020 E 2021                      #
################################################################################

# Preparar df_trafego20: selecionar e renomear colunas-chave
df_trafego20_comparativo <- df_trafego20 %>%
  select(
    ID,
    Ano,
    UF = sg_uf,
    CODIGO_BR = vl_br,
    EXTENSAO = vl_extensa,
    SUPERFICIE = ds_superfi,
    JURISDICAO = ds_jurisdi,
    TIPO_LINK = nm_tipo_tr,
    CLASSIFICACAO = classificação,
    GEH,
    VMDA_C = VMDa_C,
    VMDA_D = VMDa_D,
    VMDA_TOTAL
  ) %>%
  mutate(
    REGIAO = NA_character_,
    RELEVO_PRE = NA_character_,
    VELOCIDADE = NA_integer_,
    .after = UF
  )

# Preparar df_trafego21: selecionar colunas equivalentes
df_trafego21_comparativo <- df_trafego21 %>%
  select(
    ID = ID_PNCT,
    Ano,
    UF,
    REGIAO,
    CODIGO_BR,
    EXTENSAO,
    SUPERFICIE,
    JURISDICAO,
    TIPO_LINK,
    RELEVO_PRE,
    VELOCIDADE,
    CLASSIFICACAO,
    GEH,
    VMDA_C = VMDA_AB,
    VMDA_D = VMDA_BA,
    VMDA_TOTAL
  )

# Unificar os datasets 2020 e 2021
df_trafego_2020_2021 <- bind_rows(df_trafego20_comparativo, df_trafego21_comparativo)

cat("\n=== DATASET UNIFICADO (2020-2021) PARA ANÁLISE COMPARATIVA ===\n")
cat("Dimensões:", nrow(df_trafego_2020_2021), "linhas x", 
    ncol(df_trafego_2020_2021), "colunas\n\n")

# Distribuição por ano
cat("Distribuição de registros por Ano:\n")
print(table(df_trafego_2020_2021$Ano))

cat("\n=== ESTRUTURA DO DATASET UNIFICADO ===\n")
str(df_trafego_2020_2021)

################################################################################
#              ANÁLISES DESCRITIVAS COMPARATIVAS 2020 vs 2021                  #
################################################################################

# Resumo por ano
cat("\n=== RESUMO POR ANO ===\n")
df_trafego_2020_2021 %>%
  group_by(Ano) %>%
  summarise(
    Total_Registros = n(),
    Extensao_Media = mean(EXTENSAO, na.rm = TRUE),
    Extensao_Total = sum(EXTENSAO, na.rm = TRUE),
    VMDA_C_Media = mean(VMDA_C, na.rm = TRUE),
    VMDA_D_Media = mean(VMDA_D, na.rm = TRUE),
    VMDA_TOTAL_Media = mean(VMDA_TOTAL, na.rm = TRUE),
    VMDA_TOTAL_Mediana = median(VMDA_TOTAL, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  print()

################################################################################
#           COMPARAÇÃO POR UNIDADE DA FEDERAÇÃO (2020 vs 2021)                 #
################################################################################

cat("\n=== DISTRIBUIÇÃO POR UF - COMPARATIVO 2020 vs 2021 ===\n")
df_comparativo_uf <- df_trafego_2020_2021 %>%
  group_by(Ano, UF) %>%
  summarise(
    Total = n(),
    Extensao_Total = sum(EXTENSAO, na.rm = TRUE),
    VMDA_TOTAL_Media = mean(VMDA_TOTAL, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  pivot_wider(
    names_from = Ano,
    values_from = c(Total, Extensao_Total, VMDA_TOTAL_Media),
    names_sep = "_"
  ) %>%
  arrange(desc(Total_2021))

print(df_comparativo_uf)

################################################################################
#        COMPARAÇÃO POR CLASSIFICAÇÃO (2020 vs 2021)                           #
################################################################################

cat("\n=== DISTRIBUIÇÃO POR CLASSIFICAÇÃO - COMPARATIVO 2020 vs 2021 ===\n")
df_comparativo_classificacao <- df_trafego_2020_2021 %>%
  group_by(Ano, CLASSIFICACAO) %>%
  summarise(
    Total = n(),
    Extensao_Total = sum(EXTENSAO, na.rm = TRUE),
    VMDA_TOTAL_Media = mean(VMDA_TOTAL, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  pivot_wider(
    names_from = Ano,
    values_from = c(Total, Extensao_Total, VMDA_TOTAL_Media),
    names_sep = "_"
  ) %>%
  arrange(desc(Total_2021))  # Remover 'na.rm = TRUE'

print(df_comparativo_classificacao)

################################################################################
#        COMPARAÇÃO POR TIPO DE SUPERFÍCIE (2020 vs 2021)                      #
################################################################################

cat("\n=== DISTRIBUIÇÃO POR SUPERFÍCIE - COMPARATIVO 2020 vs 2021 ===\n")
df_comparativo_superficie <- df_trafego_2020_2021 %>%
  group_by(Ano, SUPERFICIE) %>%
  summarise(
    Total = n(),
    Extensao_Total = sum(EXTENSAO, na.rm = TRUE),
    VMDA_TOTAL_Media = mean(VMDA_TOTAL, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  pivot_wider(
    names_from = Ano,
    values_from = c(Total, Extensao_Total, VMDA_TOTAL_Media),
    names_sep = "_"
  ) %>%
  arrange(desc(Total_2021))

print(df_comparativo_superficie)

################################################################################
#         ESTATÍSTICAS DE VMDA_TOTAL POR ANO                                   #
################################################################################

cat("\n=== ESTATÍSTICAS DESCRITIVAS DE VMDA_TOTAL ===\n")
df_trafego_2020_2021 %>%
  group_by(Ano) %>%
  summarise(
    N = n(),
    N_NA = sum(is.na(VMDA_TOTAL)),
    Media = mean(VMDA_TOTAL, na.rm = TRUE),
    Mediana = median(VMDA_TOTAL, na.rm = TRUE),
    DesvPad = sd(VMDA_TOTAL, na.rm = TRUE),
    Min = min(VMDA_TOTAL, na.rm = TRUE),
    Q1 = quantile(VMDA_TOTAL, 0.25, na.rm = TRUE),
    Q3 = quantile(VMDA_TOTAL, 0.75, na.rm = TRUE),
    Max = max(VMDA_TOTAL, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  print()

################################################################################
#          VISUALIZAÇÕES COMPARATIVAS 2020 vs 2021                             #
################################################################################

# Gráfico 1: Distribuição de VMDA_TOTAL por ano
cat("\n=== GRÁFICO 1: Distribuição de VMDA_TOTAL (2020 vs 2021) ===\n")
df_trafego_2020_2021 %>%
  ggplot(aes(x = VMDA_TOTAL, fill = as.factor(Ano))) +
  geom_histogram(bins = 40, alpha = 0.7, color = "black") +
  scale_x_log10() +
  scale_fill_manual(
    name = "Ano",
    values = c("2020" = "steelblue", "2021" = "coral")
  ) +
  labs(
    title = "Distribuição de VMDA_TOTAL (Volume Médio Diário Anualizado)",
    x = "VMDA_TOTAL (escala logarítmica)",
    y = "Frequência",
    subtitle = "Comparação entre 2020 (SNV) e 2021 (PNCT)"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold", size = 12),
    plot.subtitle = element_text(size = 10, color = "gray50")
  )

# Gráfico 2: Boxplot comparativo por ano
cat("\nGRÁFICO 2: Boxplot de VMDA_TOTAL por Ano\n")
df_trafego_2020_2021 %>%
  ggplot(aes(x = as.factor(Ano), y = VMDA_TOTAL, fill = as.factor(Ano))) +
  geom_boxplot(alpha = 0.7) +
  scale_y_log10() +
  scale_fill_manual(
    name = "Ano",
    values = c("2020" = "steelblue", "2021" = "coral")
  ) +
  labs(
    title = "Comparação de VMDA_TOTAL entre 2020 e 2021",
    x = "Ano",
    y = "VMDA_TOTAL (escala logarítmica)",
    subtitle = "Distribuição com destaque para mediana, quartis e outliers"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold", size = 12),
    plot.subtitle = element_text(size = 10, color = "gray50")
  )

# Gráfico 3: Comparativo por UF (top 10)
cat("\nGRÁFICO 3: Registros por UF (2020 vs 2021)\n")
df_trafego_2020_2021 %>%
  group_by(Ano, UF) %>%
  summarise(Total = n(), .groups = 'drop') %>%
  pivot_wider(names_from = Ano, values_from = Total, values_fill = 0) %>%
  arrange(desc(`2021`)) %>%
  head(10) %>%
  pivot_longer(cols = c(`2020`, `2021`), names_to = "Ano", values_to = "Total") %>%
  ggplot(aes(x = reorder(UF, Total), y = Total, fill = Ano)) +
  geom_col(position = "dodge", alpha = 0.8) +
  coord_flip() +
  scale_fill_manual(
    values = c("2020" = "steelblue", "2021" = "coral")
  ) +
  labs(
    title = "Top 10 UF - Comparativo de Registros (2020 vs 2021)",
    x = "Unidade da Federação",
    y = "Número de Registros",
    fill = "Ano"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold", size = 12),
    plot.subtitle = element_text(size = 10, color = "gray50")
  )

# Gráfico 4: Extensão média por UF (top 10)
cat("\nGRÁFICO 4: Extensão Média por UF (2020 vs 2021)\n")
df_trafego_2020_2021 %>%
  group_by(Ano, UF) %>%
  summarise(
    Extensao_Media = mean(EXTENSAO, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  pivot_wider(names_from = Ano, values_from = Extensao_Media) %>%
  arrange(desc(`2021`)) %>%
  head(10) %>%
  pivot_longer(
    cols = c(`2020`, `2021`),
    names_to = "Ano",
    values_to = "Extensao_Media"
  ) %>%
  ggplot(aes(x = reorder(UF, Extensao_Media), y = Extensao_Media, fill = Ano)) +
  geom_col(position = "dodge", alpha = 0.8) +
  coord_flip() +
  scale_fill_manual(values = c("2020" = "steelblue", "2021" = "coral")) +
  labs(
    title = "Extensão Média de Trechos por UF (2020 vs 2021)",
    x = "Unidade da Federação",
    y = "Extensão Média (km)",
    fill = "Ano"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold", size = 12)
  )

################################################################################
#         ANÁLISE DE COBERTURA E REPRESENTATIVIDADE                            #
################################################################################

cat("\n=== COBERTURA GEOGRÁFICA E REPRESENTATIVIDADE ===\n")

# UFs presentes em 2020 vs 2021
uf_2020 <- unique(df_trafego20_comparativo$UF)
uf_2021 <- unique(df_trafego21_comparativo$UF)

cat("UFs em 2020:", length(uf_2020), "unidades\n")
cat("UFs em 2021:", length(uf_2021), "unidades\n")
cat("UFs em comum:", length(intersect(uf_2020, uf_2021)), "unidades\n")
cat("UFs apenas em 2020:", length(setdiff(uf_2020, uf_2021)), "unidades\n")
cat("UFs apenas em 2021:", length(setdiff(uf_2021, uf_2020)), "unidades\n\n")

# Listar diferenças
if (length(setdiff(uf_2020, uf_2021)) > 0) {
  cat("UFs exclusivas de 2020:", paste(setdiff(uf_2020, uf_2021), collapse = ", "), "\n")
}
if (length(setdiff(uf_2021, uf_2020)) > 0) {
  cat("UFs exclusivas de 2021:", paste(setdiff(uf_2021, uf_2020), collapse = ", "), "\n")
}
```
