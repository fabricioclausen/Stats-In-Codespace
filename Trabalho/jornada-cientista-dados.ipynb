```r
# Carregar bibliotecas necessárias
library(tidyverse)
library(readr)

# Ler o CSV com separador ponto-e-vírgula
df_trafego20 <- read.csv2(
  "Trabalho/bases-de-dados/contagem-de-trafego/contagem_de_trafego_cgplan_dez20.csv",
  sep = ";",
  encoding = "UTF-8",
  stringsAsFactors = FALSE
)

# Exibir dimensões e primeiras linhas
cat("Dimensões do conjunto de dados:", nrow(df_trafego20), "linhas x", ncol(df_trafego20), "colunas\n")

# Quantas colunas possui a tabela?
print(paste("A tabela possui ", ncol(df_trafego20), " colunas."))

# Quantas colunas possuem valores ausentes?
print(paste("A tabela possui ", sum(colSums(is.na(df_trafego20)) > 0), " colunas com valores ausentes."))

head(df_trafego20, 5)

print("O dicionário de dados ")
```

```r
# Estrutura e tipos de dados
str(df_trafego20)

# Estatísticas descritivas das variáveis numéricas
summary(df_trafego20)
```

```r
library(kableExtra)

# Distribuição de UF (Unidades da Federação)
df_trafego20 %>%
    group_by(sg_uf) %>%
    summarise(contagem = n(), .groups = 'drop') %>%
    arrange(desc(contagem)) %>%
    kable(caption = "Distribuição de observações por UF") %>%
    kable_styling(bootstrap_options = "striped")

# Visualização gráfica
df_trafego20 %>%
    group_by(sg_uf) %>%
    summarise(contagem = n(), .groups = 'drop') %>%
    arrange(desc(contagem)) %>%
    ggplot(aes(x = reorder(sg_uf, -contagem), y = contagem, fill = sg_uf)) +
    geom_col(show.legend = FALSE) +
    labs(title = "Distribuição de Contagem de (linhas) Tráfego por UF (dez/2020)",
             x = "Unidade da Federação",
             y = "Número de Registros") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
head(df_trafego20$classificação, 5)

# Mostrar a distribuição da variável 'classificação'
df_trafego20 %>%
    group_by(classificação) %>%
    summarise(contagem = n(), .groups = 'drop') %>%
    arrange(desc(contagem)) %>%
    kable(caption = "Distribuição de observações por Classificação") %>%
    kable_styling(bootstrap_options = "striped")
# Visualização gráfica
df_trafego20 %>%
    group_by(classificação) %>%
    summarise(contagem = n(), .groups = 'drop') %>%
    arrange(desc(contagem)) %>%
    ggplot(aes(x = reorder(classificação, -contagem), y = contagem, fill = classificação)) +
    geom_col(show.legend = FALSE) +
    labs(title = "Distribuição de Contagem de (linhas) Tráfego por Classificação (dez/2020)",
             x = "Classificação",
             y = "Número de Registros") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
# Converter VMDa_C para numérico
df_trafego20$VMDa_C <- as.numeric(df_trafego20$VMDa_C)

head(df_trafego20$VMDa_C, 5)

# Estatísticas descritivas
summary(df_trafego20$VMDa_C)

# Visualização gráfica com escala logarítmica no eixo Y
df_trafego20 %>%
    ggplot(aes(x = VMDa_C)) +
    geom_histogram(bins = 30, fill = "steelblue", color = "black", alpha = 0.7) +
    scale_y_log10() +
    labs(title = "Distribuição de VMDa_C - Sentido Crescente (dez/2020)",
         x = "Volume Médio Diário Anualizado (VMDa_C)",
         y = "Frequência (escala log)") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
# Preciso adicionar coluna Ano = 2020 no dataframe df_trafego20 
df_trafego20$Ano <- 2020
# Preciso salvar o dataframe atualizado com a nova coluna em um formato parquet, 
# que é mais eficiente para grandes volumes de dados.
library(arrow)

# Exemplo: particionar por coluna "ano"
write_dataset(
  df_trafego20,
  path = "Trabalho/bases-de-dados/contagem-de-trafego/contagem_de_trafego_cgplan_dez20",
  format = "parquet",
  partitioning = "Ano"
)
```
