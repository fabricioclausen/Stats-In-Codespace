```r
# gerado na maquina do professor


################################################################################
#                 CARREGAR DADOS DE TRÁFEGO 2022 - SNV 202301B                 #
################################################################################

# Carregar bibliotecas necessárias
library(tidyverse)
library(readr)

# Ler o CSV com separador ponto-e-vírgula
# Nota: O arquivo contém valores vazios (-) e espaços que precisam ser tratados
# Similar ao padrão 2020, mas com coluna adicional 'ul'
df_trafego22 <- read.csv2(
  "Trabalho/bases-de-dados/contagem-de-trafego/contagem_de_trafego_cgplan_jan23.csv",
  sep = ";",
  encoding = "UTF-8",
  stringsAsFactors = FALSE,
  na.strings = c("", "-", " ")  # Tratar vazio, hífen e espaço como NA
)

# Exibir dimensões e primeiras linhas
cat("\n=== RESUMO DO CARREGAMENTO (2022 - SNV 202301B) ===\n")
cat("Dimensões do conjunto de dados:", 
    nrow(df_trafego22), "linhas x", 
    ncol(df_trafego22), "colunas\n\n")

# Quantas colunas possui a tabela?
print(paste("A tabela possui", ncol(df_trafego22), "colunas."))

# Quantas colunas possuem valores ausentes?
colunas_com_na <- sum(colSums(is.na(df_trafego22)) > 0)
print(paste("A tabela possui", colunas_com_na, "colunas com valores ausentes."))

# Mostrar primeiras linhas
cat("\n=== PRIMEIRAS 6 LINHAS ===\n")
print(head(df_trafego22, 6))

################################################################################
#                      ESTRUTURA E TIPOS DE DADOS                             #
################################################################################

# Estrutura do dataframe
cat("\n=== ESTRUTURA DOS DADOS ===\n")
str(df_trafego22)

# Estatísticas descritivas das variáveis numéricas
cat("\n=== ESTATÍSTICAS DESCRITIVAS ===\n")
summary(df_trafego22)

################################################################################
#              CONVERSÃO E LIMPEZA DE COLUNAS NUMÉRICAS (2022)                 #
################################################################################

# Converter colunas de volume para numérico
# As colunas com volumes (A_C até J_D e VMDa_*)
colunas_volume <- c(
  paste0(c("A", "B", "C", "D", "E", "F", "G", "H", "I", "J"), "_C"),
  paste0(c("A", "B", "C", "D", "E", "F", "G", "H", "I", "J"), "_D"),
  "VMDa_C", "VMDa_D"
)

# Aplicar conversão com tratamento de pontos decimais
for (col in colunas_volume) {
  if (col %in% names(df_trafego22)) {
    # Substituir vírgula por ponto e converter para numérico
    df_trafego22[[col]] <- as.numeric(
      gsub(",", ".", gsub("\\.", "", df_trafego22[[col]]))
    )
  }
}

# Converter vl_extensa para numérico (também pode ter vírgula)
df_trafego22$vl_extensa <- as.numeric(
  gsub(",", ".", gsub("\\.", "", df_trafego22$vl_extensa))
)

cat("\n=== APÓS CONVERSÃO DE TIPOS (2022) ===\n")
str(df_trafego22)

################################################################################
#                    ANÁLISES EXPLORATÓRIAS INICIAIS (2022)                    #
################################################################################

# Distribuição por UF
cat("\n=== DISTRIBUIÇÃO POR UF (2022) ===\n")
df_trafego22 %>%
  group_by(sg_uf) %>%
  summarise(
    Total = n(),
    .groups = 'drop'
  ) %>%
  arrange(desc(Total)) %>%
  print()

# Distribuição por classificação
cat("\n=== DISTRIBUIÇÃO POR CLASSIFICAÇÃO (2022) ===\n")
df_trafego22 %>%
  group_by(classificação) %>%
  summarise(
    Total = n(),
    .groups = 'drop'
  ) %>%
  arrange(desc(Total)) %>%
  print()

# Estatísticas descritivas de VMDa_C
cat("\n=== ESTATÍSTICAS VMDa_C - SENTIDO CRESCENTE (2022) ===\n")
summary(df_trafego22$VMDa_C)

# Estatísticas descritivas de VMDa_D
cat("\n=== ESTATÍSTICAS VMDa_D - SENTIDO DECRESCENTE (2022) ===\n")
summary(df_trafego22$VMDa_D)

################################################################################
#         COMPARAÇÃO ENTRE 2020, 2021 E 2022 - ESTRUTURA DE COLUNAS           #
################################################################################

# 2020 (df_trafego20) usa: A_C, B_C, ... J_C (Crescente) + A_D, B_D, ... J_D (Decrescente)
# 2021 (df_trafego21) usa: A_AB, B_AB, ... J_AB (AB) + A_BA, B_BA, ... J_BA (BA)
# 2022 (df_trafego22) usa: A_C, B_C, ... J_C (Crescente) + A_D, B_D, ... J_D (Decrescente)

cat("\n=== COMPATIBILIDADE DE COLUNAS ===\n")
cat("2020 (SNV 202001A) - Formato Crescente/Decrescente\n")
cat("2021 (PNCT) - Formato AB/BA\n")
cat("2022 (SNV 202301B) - Formato Crescente/Decrescente (compatível com 2020)\n")

################################################################################
#   UNIFICAR DATASETS 2020, 2021 E 2022 EM ARQUIVO PARQUET PARTICIONADO       #
################################################################################

# Preparar dataset 2020: renomear colunas de sentidos
df_trafego20_prep <- df_trafego20 %>%
  rename(
    A_AB = A_C, B_AB = B_C, C_AB = C_C, D_AB = D_C, E_AB = E_C,
    F_AB = F_C, G_AB = G_C, H_AB = H_C, I_AB = I_C, J_AB = J_C,
    A_BA = A_D, B_BA = B_D, C_BA = C_D, D_BA = D_D, E_BA = E_D,
    F_BA = F_D, G_BA = G_D, H_BA = H_D, I_BA = I_D, J_BA = J_D,
    VMDA_AB = VMDa_C,
    VMDA_BA = VMDa_D,
    UF = sg_uf,
    CLASSIFICACAO = classificação
  ) %>%
  select(
    ID, CODIGO_BR = vl_br, UF, 
    EXTENSAO = vl_extensa, SUPERFICIE = ds_superfi,
    JURISDICAO = ds_jurisdi, TIPO_LINK = nm_tipo_tr,
    CLASSIFICACAO, GEH,
    A_AB, B_AB, C_AB, D_AB, E_AB, F_AB, G_AB, H_AB, I_AB, J_AB,
    A_BA, B_BA, C_BA, D_BA, E_BA, F_BA, G_BA, H_BA, I_BA, J_BA,
    VMDA_AB, VMDA_BA
  ) %>%
  mutate(
    REGIAO = NA_character_,
    RELEVO_PRE = NA_character_,
    VELOCIDADE = NA_integer_,
    VMDA_TOTAL = VMDA_AB + VMDA_BA,
    Ano = 2020,
    .before = VMDA_AB
  )

# Preparar dataset 2021
df_trafego21_prep <- df_trafego21 %>%
  mutate(
    Ano = 2021,
    ID = ID_PNCT
  ) %>%
  select(
    ID, CODIGO_BR, UF, REGIAO, 
    EXTENSAO, SUPERFICIE, JURISDICAO, TIPO_LINK,
    RELEVO_PRE, VELOCIDADE, CLASSIFICACAO, GEH,
    A_AB, B_AB, C_AB, D_AB, E_AB, F_AB, G_AB, H_AB, I_AB, J_AB,
    A_BA, B_BA, C_BA, D_BA, E_BA, F_BA, G_BA, H_BA, I_BA, J_BA,
    VMDA_AB, VMDA_BA, VMDA_TOTAL, Ano
  )

# Preparar dataset 2022: similar ao 2020
df_trafego22_prep <- df_trafego22 %>%
  rename(
    A_AB = A_C, B_AB = B_C, C_AB = C_C, D_AB = D_C, E_AB = E_C,
    F_AB = F_C, G_AB = G_C, H_AB = H_C, I_AB = I_C, J_AB = J_C,
    A_BA = A_D, B_BA = B_D, C_BA = C_D, D_BA = D_D, E_BA = E_D,
    F_BA = F_D, G_BA = G_D, H_BA = H_D, I_BA = I_D, J_BA = J_D,
    VMDA_AB = VMDa_C,
    VMDA_BA = VMDa_D,
    UF = sg_uf,
    CLASSIFICACAO = classificação
  ) %>%
  select(
    ID, CODIGO_BR = vl_br, UF, 
    EXTENSAO = vl_extensa, SUPERFICIE = ds_superfi,
    JURISDICAO = ds_jurisdi, TIPO_LINK = nm_tipo_tr,
    CLASSIFICACAO, GEH,
    A_AB, B_AB, C_AB, D_AB, E_AB, F_AB, G_AB, H_AB, I_AB, J_AB,
    A_BA, B_BA, C_BA, D_BA, E_BA, F_BA, G_BA, H_BA, I_BA, J_BA,
    VMDA_AB, VMDA_BA
  ) %>%
  mutate(
    REGIAO = NA_character_,
    RELEVO_PRE = NA_character_,
    VELOCIDADE = NA_integer_,
    VMDA_TOTAL = VMDA_AB + VMDA_BA,
    Ano = 2022,
    .before = VMDA_AB
  )

# Unificar os três datasets
df_trafego_unificado <- bind_rows(df_trafego20_prep, df_trafego21_prep, df_trafego22_prep)

cat("\n=== DATASET UNIFICADO (2020-2021-2022) ===\n")
cat("Dimensões:", nrow(df_trafego_unificado), "linhas x", 
    ncol(df_trafego_unificado), "colunas\n")
cat("Período: 2020 a 2022\n\n")

# Distribuição por ano
cat("Distribuição por Ano:\n")
print(table(df_trafego_unificado$Ano))

cat("\n=== PRIMEIRAS 3 LINHAS (AMOSTRA) ===\n")
print(head(df_trafego_unificado, 3))

# Salvar em Parquet particionado por Ano
library(arrow)

write_dataset(
  df_trafego_unificado,
  path = "Trabalho/bases-de-dados/contagem-de-trafego/trafego_unificado_2020_2021_2022",
  format = "parquet",
  partitioning = "Ano"
)

cat("\n✓ Dataset unificado (2020-2021-2022) salvo com particionamento por Ano\n")
cat("   Caminho: Trabalho/bases-de-dados/contagem-de-trafego/trafego_unificado_2020_2021_2022/\n")

################################################################################
#              LEITURA DO DATASET PARTICIONADO UNIFICADO                       #
################################################################################

# Abrir dataset particionado
df_trafego_unif_parquet <- open_dataset(
  "Trabalho/bases-de-dados/contagem-de-trafego/trafego_unificado_2020_2021_2022/"
)

cat("\n=== VERIFICAÇÃO DO DATASET PARTICIONADO ===\n")
cat("Schema do dataset:\n")
print(df_trafego_unif_parquet$schema)

# Consultar amostra por ano
cat("\n=== AMOSTRA POR ANO ===\n")
for (ano in c(2020, 2021, 2022)) {
  amostra <- df_trafego_unif_parquet %>%
    filter(Ano == ano) %>%
    head(2) %>%
    collect()
  
  cat("\nAno", ano, "- Primeiros 2 registros:\n")
  print(amostra)
}
```
